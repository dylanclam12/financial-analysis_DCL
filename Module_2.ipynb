## Section 1: Sentiment Analysis of APPL Q4 2024 Earnings Call Transcript

In this module, we will be completing a sentiment analysis of the recent APPL Q4 2024 earnings call. We will be using the [NLTK (Natural Language Toolkit)](https://www.nltk.org/) Python package. The SentimentIntensityAnalyzer from the VADER (Valence Aware Dictionary and sEntiment Reasoner) package is a tool used to analyze the sentiment of text. It is specifically designed to work well with social media text, such as tweets, which may contain slang, abbreviations, and informal expressions.

- Sentiment Scores


    - Positive (pos): A measure of how positive the sentiment is.
    - Negative (neg): A measure of how negative the sentiment is.
	- Neutral (neu): A measure of how neutral the sentiment is.
	- Compound (compound): A normalized score that represents the overall sentiment, ranging from -1 (most negative) to +1 (most positive)
	    - Positive: Compound score > 0.05
        - Neutral: -0.05 <= Compound score <= 0.05
        - Negative: Compound score < -0.05

- Lexicon-Based Approach

    - VADER works by analyzing the words in the text based on a pre-built lexicon that assigns scores to words based on their sentiment (positive, negative, or neutral). It also accounts for:
        - Punctuation: The tool considers punctuation, such as exclamation points and question marks, which can indicate the intensity of sentiment.
        - Negations: VADER handles negations (e.g., “not good” is negative) by adjusting the sentiment scores accordingly.
        - Emojis and Slang: It has specific rules for handling common slang and emojis, which makes it effective for analyzing spoken text.


In this section We will:


- Separate the earnings call transcript into individual lines.
- Preprocess the text by filtering out stopwords that provide little evidence of sentiment.
- Tokenize words in each line.
- Derive the sentiment of each line using the VADER SentimentIntensityAnalyzer.

```python
sentiment_df = fa.sn.sentiment_analysis("text/APPL_Q4_Earnings_Call.txt")
sentiment_df
```

```python
avg_compound = round(sentiment_df["compound"].mean(), 4)
print(f"Average Compound Score = {avg_compound}")
```

>**Since the average compound score is 0.6235 which is greater than 0.05, the earnings call has a more positive sentiment.**
