### Security Market Line (SML)
The **Securities Market Line (SML)** is a graphical representation of the **Capital Asset Pricing Model (CAPM)**, which describes the relationship between the expected return of an asset and its systematic risk (beta, $\beta$).
The SML is a straight line that plots **Expected Return ($E(R_i)$)** on the vertical axis and **Systematic Risk (Beta, $\beta$)** on the horizontal axis. It shows how much return investors require for taking on a certain level of market risk.
The slope of the SML is the **market risk premium** which represents the additional return expected for taking on market risk over the risk-free rate: 
$$
E(R_m) - R_f
$$

Securities Market Line (SML) Equation:
$$
E(R_i) = R_f + \beta_i \cdot [E(R_m) - R_f]
$$

Variables:
- **$E(R_i)$:** Expected return of the security $i$.
- **$R_f$:** Risk-free rate of return (e.g., returns on government bonds).
- **$\beta_i$:** Beta of the security $i$, a measure of its systematic risk relative to the market.
- **$E(R_m)$:** Expected return of the market portfolio.
- **$E(R_m) - R_f$:** Market risk premium.

### Alpha Calculation
The **alpha ($\alpha$)** of a stock measures the excess return of the stock beyond what is predicted by the CAPM. It quantifies whether a security outperformed or underperformed relative to the expected return based on its beta.

The alpha equation is given by:
$$
\alpha_i = R_i - \left[R_f + \beta_i \cdot (E(R_m) - R_f)\right]
$$

Variables:
- **$\alpha_i$:** Alpha of the security $i$, representing the excess return.
- **$R_i$:** Actual return of the security $i$.
- **$R_f$:** Risk-free rate of return.
- **$\beta_i$:** Beta of the security $i$, a measure of its systematic risk.
- **$E(R_m)$:** Expected return of the market portfolio.
- **$E(R_m) - R_f$:** Market risk premium.

Interpretation:
- A **positive alpha ($\alpha > 0$)** indicates that the stock outperformed its expected return.
- A **negative alpha ($\alpha < 0$)** indicates underperformance relative to its expected return.
